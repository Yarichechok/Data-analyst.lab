### Завдання 8

```python
def outlier_cleaner_IQR(series):
    q1, q3 = series.quantile([0.25, 0.75])
    iqr = q3 - q1
    low, high = q1 - 1.5*iqr, q3 + 1.5*iqr
    mask = (series < low) | (series > high)
    s = series.copy()
    s[mask] = np.nan
    return s
```

### Завдання 9

```python
width_clear_IQR = outlier_cleaner_IQR(df.width)
```

### Завдання 10

```python
width_clear_IQR.isna().sum()
```

### Завдання 11

### Завдання 12

```python
sns.boxplot(data=width_clear_IQR.dropna())
```

### Завдання 13

```python
def outlier_cleaner_median(series):
    med = series.median()
    mad = (series - med).abs().median()
    s = series.copy()
    s[(series - med).abs() > 3*mad] = pd.NA
    return s

width_clear_median = outlier_cleaner_median(df.width)
width_clear_median.isna().sum()
```

```python
sns.stripplot(y=df["width"], hue=width_clear_median.isna()); plt.show()
```

```python
sns.boxplot(x=width_clear_median.dropna()); plt.show()
```

### Завдання 14

```python
def outlier_cleaner_zscore(series):
    s = series.copy()
    z = pd.Series(zscore(series), index=series.index).abs()
    s[z > 3] = pd.NA
    return s

width_clear_zscore = outlier_cleaner_zscore(df.width)
width_clear_zscore.isna().sum()

```

```python
sns.stripplot(y=df["width"], hue=width_clear_zscore.isna()); plt.show()
```

```python
sns.boxplot(x=width_clear_zscore.dropna()); plt.show()
```
